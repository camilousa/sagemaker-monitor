```python
import boto3
import numpy as np
```

```python
%store -r endpoint_name
```

```python
client = boto3.client("sagemaker-runtime", "us-east-1")
```

```python
def generate_payload():
    age = np.random.randint(10, 90)
    sex = np.random.choice([0, 1])
    chestpain = np.random.choice([0, 1, 2, 3])
    blood = np.random.randint(94, 200)
    chol = np.random.randint(94, 500)
    payload = f"{age},{sex},{chestpain},{blood},{chol},0,1,123,1,0.2,1,0,3\n"
    return payload
```

```python
num = 100
while True:
    for id in range(num):
            payload = generate_payload()
            response = client.invoke_endpoint(
               EndpointName = endpoint_name,
               Body=payload,
               ContentType="text/csv",
               Accept="text/csv", 
               InferenceId=f"{id}" 
            )
            #print(response["Body"].read())
```
